```python
# Strip any leading or trailing spaces from column names
df.columns = df.columns.str.strip()

# Convert InvoiceDate to datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Convert numerical columns
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'].replace(r'[\$,]', '', regex=True), errors='coerce')
df['TotalCost'] = pd.to_numeric(df['TotalCost'].replace(r'[\$,]', '', regex=True), errors='coerce')

# Check if conversion was successful
print(df.dtypes)
```

```python
from IPython.display import display, HTML


# 1. Ordinal Encoding
priority_mapping = {'High': 3, 'Medium': 2, 'Low': 1}
satisfaction_mapping = {
    'Very Satisfied': 4,
    'Satisfied': 3,
    'Dissatisfied': 2,
    'Very Dissatisfied': 1
}

df['OrderPriority'] = df['OrderPriority'].map(priority_mapping).astype('Int64')
df['CustomerOrderSatisfaction'] = df['CustomerOrderSatisfaction'].map(satisfaction_mapping).astype('Int64')

# 2. Nominal One-Hot Encoding
df_nominal = pd.get_dummies(df[['Region', 'DiscountApplied']], drop_first=False)

# 3. Combine
df_encoded = pd.concat([df, df_nominal], axis=1)
df_encoded = df_encoded.drop(['Region', 'DiscountApplied'], axis=1)

# 4. Display
display(HTML(df_encoded.head().to_html(index=False)))

# 5. Save the encoded DataFrame to a xlsx file
df_encoded.to_excel('encoded_dataset.xlsx', index=False)

```

```python
from mlxtend.preprocessing import TransactionEncoder

# Step 1: Group products by OrderID
basket = df.groupby('OrderID')['ProductName'].apply(list).tolist()

# Step 2: One-hot encode using TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(basket).transform(basket)
df_basket = pd.DataFrame(te_ary, columns=te.columns_)

# Display the encoded basket
display(df_basket.head())

# Export MBA df to xlsx file
df_basket.to_excel('basket_encoded.xlsx', index=False)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
from IPython.display import display, HTML

# Apply the apriori algorithm
frequent_itemsets = apriori(df_basket, min_support=0.01, use_colnames=True)

# Generate the association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Display the rules in an HTML table format
display(HTML(rules.head().to_html(index=False)))
```

```python
#Confidence and lift are the two best support metrics to use for association rules.

#Confidence measures how often items in the rule appear together in the dataset, while lift measures how much more likely the items are to appear together than if they were statistically independent.

#Sort rules table by lift
rules_sorted_lift = rules.sort_values(by='lift', ascending=False)

pd.set_option('display.max_colwidth', None)

rules_sorted_lift.head()
```
