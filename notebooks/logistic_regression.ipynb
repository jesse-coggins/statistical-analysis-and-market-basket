```python
# D600 Statistical Data Mining
# Task 2
# Logistic Regression

#Importing Libraries
import pandas as pd # pandas for data manipulation and analysis


#Importing Dataset
file_path = r"../data/housing_data.csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to understand its structure
display(df.head())
```

```python
#Descriptive statistics
variable_columns = ['NumBathrooms', 'NumBedrooms', 'IsLuxury']
stats = df[variable_columns].describe()

# Display the results
display(stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a brighter overall style
sns.set_style("whitegrid")

# Univariate: Histogram for NumBathrooms
plt.figure(figsize=(7,5))
sns.histplot(df['NumBathrooms'], bins=20, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Number of Bathrooms', fontsize=16)
plt.xlabel('Number of Bathrooms', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Univariate: Histogram for NumBedrooms
plt.figure(figsize=(7,5))
sns.histplot(df['NumBedrooms'], bins=20, kde=True, color='lightgreen', edgecolor='black')
plt.title('Distribution of Number of Bedrooms', fontsize=16)
plt.xlabel('Number of Bedrooms', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Univariate: Bar plot for IsLuxury
plt.figure(figsize=(7,5))
sns.countplot(x='IsLuxury', hue='IsLuxury', data=df, palette='Set2', edgecolor='black', legend=False)
plt.title('Distribution of Luxury Status', fontsize=16)
plt.xlabel('Is Luxury (0 = No, 1 = Yes)', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()

# Bivariate: Boxplot of NumBathrooms vs IsLuxury
plt.figure(figsize=(7,5))
sns.boxplot(x='IsLuxury', y='NumBathrooms', data=df, hue='IsLuxury', palette='coolwarm', legend=False)
plt.title('Bathrooms by Luxury Status', fontsize=16)
plt.xlabel('Luxury Status (0 = No, 1 = Yes)', fontsize=14)
plt.ylabel('Number of Bathrooms', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Bivariate: Boxplot of NumBedrooms vs IsLuxury
plt.figure(figsize=(7,5))
sns.boxplot(x='IsLuxury', y='NumBedrooms', data=df, hue='IsLuxury', palette='coolwarm', legend=False)
plt.title('Bedrooms by Luxury Status', fontsize=16)
plt.xlabel('Luxury Status (0 = No, 1 = Yes)', fontsize=14)
plt.ylabel('Number of Bedrooms', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# 1. Select only the needed variables
selected_columns = ['NumBedrooms', 'NumBathrooms', 'IsLuxury']
df_selected = df[selected_columns]

# 2. Train-test split (80% train, 20% test)
train_df, test_df = train_test_split(df_selected, test_size=0.2, random_state=42)

# 3. Save the datasets
train_df.to_csv('train_dataset.csv', index=False)
test_df.to_csv('test_dataset.csv', index=False)

print("Training and test datasets created and saved successfully.")
```

```python
import statsmodels.api as sm

# 1. Load the training data
train_df = pd.read_csv('train_dataset.csv')

# 2. Define X and y
X_train = train_df[['NumBedrooms', 'NumBathrooms']]
y_train = train_df['IsLuxury']

# 3. Add constant term (for intercept)
X_train = sm.add_constant(X_train)

# 4. Build the logistic regression model
logit_model = sm.Logit(y_train, X_train)
result = logit_model.fit()

# 5. Print full summary (Pseudo R², p-values, coefficients)
print(result.summary())

# 6. Manually print AIC and BIC
print("\nAdditional Model Information:")
print(f"AIC: {result.aic}")
print(f"BIC: {result.bic}")
```

```python
# Optimization

# 1. Build full logistic regression model
X_full = train_df[['NumBedrooms', 'NumBathrooms']]
y = train_df['IsLuxury']

X_full_const = sm.add_constant(X_full)
full_model = sm.Logit(y, X_full_const).fit()

# 2. Show full model summary
print("\nFull Model Summary:")
print(full_model.summary())

# 3. Now pretend: Let's say 'NumBedrooms' had a p-value > 0.05
# We would manually remove it and refit the model

# (Example: reduced model without 'NumBedrooms')
X_reduced = train_df[['NumBathrooms']]  # Drop NumBedrooms
X_reduced_const = sm.add_constant(X_reduced)
reduced_model = sm.Logit(y, X_reduced_const).fit()

# 4. Show reduced model summary
print("\nReduced Model Summary:")
print(reduced_model.summary())
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

# 1. Predict on the training set
# Predict probabilities
y_train_pred_probs = result.predict(X_train)

# Convert probabilities to 0 or 1 using 0.5 threshold
y_train_pred = (y_train_pred_probs >= 0.5).astype(int)

# 2. Confusion matrix
cm = confusion_matrix(y_train, y_train_pred)
print("Confusion Matrix:")
print(cm)

# 3. Accuracy
accuracy = accuracy_score(y_train, y_train_pred)
print("\nAccuracy on Training Set:", round(accuracy, 4))
```

```python
# 1. Load the test dataset
test_df = pd.read_csv('test_dataset.csv')

# 2. Prepare X_test (same two variables used: NumBedrooms and NumBathrooms)
X_test = test_df[['NumBedrooms', 'NumBathrooms']]
X_test = sm.add_constant(X_test)  # Add constant for intercept

# 3. Define y_test
y_test = test_df['IsLuxury']

# 4. Predict probabilities
y_test_pred_probs = result.predict(X_test)

# 5. Convert probabilities to 0 or 1 using 0.5 threshold
y_test_pred = (y_test_pred_probs >= 0.5).astype(int)

# 6. Confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score

cm_test = confusion_matrix(y_test, y_test_pred)
print("Confusion Matrix (Test Set):")
print(cm_test)

# 7. Accuracy
accuracy_test = accuracy_score(y_test, y_test_pred)
print("\nAccuracy on Test Set:", round(accuracy_test, 4))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Get predicted probabilities
train_pred_probs = result.predict(X_train)

# Create logit (log-odds)
logit = np.log(train_pred_probs / (1 - train_pred_probs))

# Plot NumBedrooms vs logit
plt.scatter(train_df['NumBedrooms'], logit, alpha=0.5)
plt.title('NumBedrooms vs Logit (Log-Odds)')
plt.xlabel('NumBedrooms')
plt.ylabel('Logit')
plt.grid(True)
plt.show()

# Plot NumBathrooms vs logit
plt.scatter(train_df['NumBathrooms'], logit, alpha=0.5)
plt.title('NumBathrooms vs Logit (Log-Odds)')
plt.xlabel('NumBathrooms')
plt.ylabel('Logit')
plt.grid(True)
plt.show()
```
